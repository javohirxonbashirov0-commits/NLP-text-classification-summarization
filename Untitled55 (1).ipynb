```python
from evaluate import load
import numpy as np

def compute_metrics(eval_pred):
    # Metrikalarni yuklash
    accuracy_m = load("accuracy")
    precision_m = load("precision")
    recall_m = load("recall")
    f1_m = load("f1")

    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)


    results = {}
    results.update(accuracy_m.compute(predictions=predictions, references=labels))
    results.update(precision_m.compute(predictions=predictions, references=labels, average="weighted"))
    results.update(recall_m.compute(predictions=predictions, references=labels, average="weighted"))
    results.update(f1_m.compute(predictions=predictions, references=labels, average="weighted"))

    return results
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from evaluate import load
import numpy as np
import torch

print("Dataset yuklanmoqda...")
imdb = load_dataset("imdb")

# 3. Tokenizer va Model soz.lari
model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})

# 4. Tokenizatsiya funksiyasi
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=128)

print("Tokenizatsiya qilinmoqda...")
tokenized_imdb = imdb.map(tokenize_function, batched=True)

# 5. Metrikalarni hisoblash funksiyasi
def compute_metrics(eval_pred):
    metric_acc = load("accuracy")
    metric_f1 = load("f1")

    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    return {
        "accuracy": metric_acc.compute(predictions=predictions, references=labels)["accuracy"],
        "f1": metric_f1.compute(predictions=predictions, references=labels, average="weighted")["f1"]
    }

# 6. Modelni yaratish
model_imdb = AutoModelForSequenceClassification.from_pretrained(model_checkpoint, num_labels=2)
model_imdb.resize_token_embeddings(len(tokenizer))

# 7. TrainingArguments
training_args_imdb = TrainingArguments(
    output_dir="./results_imdb",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    num_train_epochs=1,
    weight_decay=0.01,
    logging_steps=100,
    report_to="none"
)
trainer_imdb = Trainer(
    model=model_imdb,
    args=training_args_imdb,
    train_dataset=tokenized_imdb["train"].shuffle(seed=42).select(range(5000)), # Tezroq bo'lishi uchun 5k ta
    eval_dataset=tokenized_imdb["test"].select(range(500)),
    compute_metrics=compute_metrics,
)
print("O'qitish boshlandi...")
trainer_imdb.train()
```

```python
import torch
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model_imdb.to(device)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        logits = model_imdb(**inputs).logits

    prediction = torch.argmax(logits, dim=-1).item()

    return "Positive (Zo'r!)" if prediction == 1 else "Negative (Yomon...)"
my_review = "This movie was a masterpiece, I really enjoyed every second of it!"
print(f"Natija: {predict_sentiment(my_review)}")
```

```python
def predict_news_category(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model_ag.to(device)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        logits = model_ag(**inputs).logits

    prediction = torch.argmax(logits, dim=-1).item()

    # AG News klasslari tartibi
    labels = ["World (Dunyo)", "Sports (Sport)", "Business (Biznes)", "Sci/Tech (Texno)"]
    return labels[prediction]

my_news = "Nvidia's new AI chip is breaking all performance records in the industry."
print(f"Yangilik turi: {predict_news_category(my_news)}")
```
